```python
import pandas as pd

import os

folder_path = '/content/drive/My Drive/teste_marcacao/revisar/resultados'

try:
    os.chdir(folder_path)
    print(f"Diretório de trabalho atual alterado para: {os.getcwd()}")
    print("Listando conteúdo do diretório para verificação:")
    !ls -F
except FileNotFoundError:
    print(f"ERRO: O caminho '{folder_path}' não foi encontrado. Por favor, verifique e corrija o caminho.")
except Exception as e:
    print(f"Ocorreu um erro ao mudar para o diretório '{folder_path}': {e}")

file_path1 = 'conteudo_lote_15_filtrado_limpo.xlsx'  # Substitua pelo caminho do seu primeiro arquivo
file_path2 = 'conteudo_lote_15_filtrado_limpo.xlsx'  # Substitua pelo caminho do seu segundo arquivo

# Carrega as planilhas, ignorando o cabeçalho existente
df1 = pd.read_excel(file_path1, header=None)
df2 = pd.read_excel(file_path2, header=None)

print("Primeira planilha (sem cabeçalho):")
display(df1.head())
print("\nSegunda planilha (sem cabeçalho):")
display(df2.head())
```

Agora vamos juntar as duas planilhas e definir um novo cabeçalho.

```python
df_combined = pd.concat([df1, df2], ignore_index=True)

new_columns = ['link', 'conteudo']

df_combined.columns = new_columns

df_combined = df_combined[~((df_combined['link'] == 'Link') & (df_combined['conteudo'] == 'conteudo'))]

print("Planilha combinada com o novo cabeçalho e linhas indesejadas removidas:")
display(df_combined.head())
```

```python
output_file_path = 'planilha_combinada.xlsx'
df_combined.to_csv(output_file_path, index=False)
print(f"Planilha combinada salva em: {output_file_path}")
```
